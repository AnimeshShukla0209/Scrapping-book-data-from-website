```python
# Importing Different libraries which would helpl to scrap and modify the data
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
#Loading the DataFrame
df = pd.read_csv("Book_Data_From_Multiple_Pages.csv")
```

```python
df
```

```python
data = []

for link in tqdm(df['Book_link']):
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    typ = soup.find('ul', class_ = 'breadcrumb').find_all('a')[2].text
    
    temp = soup.find('table', class_ = 'table table-striped').find_all('td')
    
    upc      = temp[0].text
    price_x  = temp[2].text[2:]
    price_i  = temp[3].text[2:]
    tax      = temp[4].text[2:]
    qn       = temp[5].text
    reviews  = temp[6].text
    
    data.append([typ, price_x, price_i, tax, qn, upc, reviews])
```

```python
data = pd.DataFrame(data,columns=["Category","Price_E_Price","Price_I_Price","Tax","QN","UPC","Reviews"])
```
