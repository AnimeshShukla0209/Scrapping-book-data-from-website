```python
#importing important libraries
import pandas as pd
import numpy as np
```

```python
# Our first data which comprises of Books , Price and their stock availability 

data1 = pd.read_csv("Book_Data_From_Multiple_Pages.csv")

# Our second data which comprises of Category , Quntity , No of reviews and many moer things

data_2 = pd.read_csv("Data_2.csv")
```

```python
#Checking data1
data1.head()
```

```python
#Checking data2
data_2.head()
```

## Forming a new DataFrame which would contain combined data of both the tables


```python
df = pd.DataFrame()
```

```python
df["Title"]           = data1["Title"]
df["Category"]        = data_2["Category"]
df["Exclusive_Price"] = data_2["Price_E_Price"]
df["Inclusive_Price"] = data_2["Price_I_Price"]
df["Rating"]          = data1["Rating"]
df["Stock"]           = data1["Stock"]
for i in data_2["QN"].values:
    df["Quantity available"] = i[10:12]
df["Reviews"]         = data_2["Reviews"]
df["Book_link"]       = data1["Book_link"]
df["Img_Link"]        = data1["Img_link"]
```
